```python
import glob
import pandas as pd
import os
experiments_dir = "AllData"
pictures_dir = "edgelists/new_format/"+experiments_dir
df = pd.read_csv("data/from-azure/"+experiments_dir+"-labelled.csv")

# THIS USED FOR REPLACING MAPF ALGORITHM NAMES WITH 0-4 VALUES
df['Ylabel'] = df['Y'].apply(replace_label_with_int)
for file in list(glob.glob(pictures_dir+"/*.png")):
    if 'label' in file:
        continue
    
    if 'Instance' not in file:
#         print("Skipping",file)
        continue
    
    instance_details = file.split('\\')[1].split('-')
    gridsize = instance_details[1]
    obstacle_ratio = instance_details[2]
    num_agents = instance_details[3]
    instance_id = instance_details[4].split('.')[0]
    
    orig_filename =file.split('\\')[1]
    print(gridsize, obstacle_ratio, num_agents, instance_id, file)

    label = df[(df.NumOfAgents==int(num_agents)) & 
               (df.InstanceId == int(instance_id)) &
                (df.GridRows == int(gridsize)) &
               (df.NumOfObstacles == obstacle_ratio_to_num(int(obstacle_ratio), int(gridsize)))]['Ylabel'].values[0]
    print(label)
    new_filename = orig_filename.split('.')[0] +"-label"+ str(label)+".png"
    print(num_agents, instance_id)
    print("Rename", orig_filename, "to", new_filename)
    os.rename(pictures_dir+"/"+orig_filename, pictures_dir+"/"+new_filename)
#     print(df[(df['NumOfAgents']==num_agents) & (df['InstanceId']==instance_id)]['Y'])
```

```python
def round_up_to_5(num):
    return str(5 - (int(num) % 5) + int(num))
```

```python
def obstacle_density_to_num(density):
    if density=='0.0':
        return 0
    density = density.split('.')[1]
    if len(density) > 2:
        return round_up_to_5(density[:2])
    elif density[0] == '0':
        return density[1:]
    elif len(density) == 1:
        return density+'0'

    return density
obstacle_density_to_num('0.10')
```

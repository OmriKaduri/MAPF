```python
from h2o.automl import H2OAutoML
aml = H2OAutoML(max_models = 20, max_runtime_secs=200, seed = 1, balance_classes=True)
```

```python
aml.train(x=features_cols, y='Y', training_frame=mapf_train, validation_frame=mapf_test)
```
